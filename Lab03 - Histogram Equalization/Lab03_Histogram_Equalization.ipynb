```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('im_cells.png')

# Verificar se a imagem foi carregada corretamente
if img is None:
    print('Não foi possível carregar a imagem.')
    exit()

# Inverter as cores para criar a imagem negativa
negative_img = 255 - img

# Inverter as cores novamente para obter a imagem original
original_img = 255 - negative_img

# Configurar a exibição das imagens lado a lado
plt.figure(figsize=(12, 6))

# Exibir a imagem original
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')
plt.axis('off')

# Exibir a imagem negativa
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(negative_img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Negativa')
plt.axis('off')

# Exibir a imagem recuperada (original a partir da imagem negativa)
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Recuperada')
plt.axis('off')

# Mostrar as imagens
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('im_cells.png', cv2.IMREAD_GRAYSCALE)

# Verificar se a imagem foi carregada corretamente
if img is None:
    print('Não foi possível carregar a imagem.')
    exit()

# Constante de ajuste para controlar o aumento do contraste
c = 0.5

# Aplicar a transformação logarítmica
log_transformed_img = c * np.log1p(img.astype(np.float32))

# Normalizar para o intervalo [0, 255]
log_transformed_img = cv2.normalize(log_transformed_img, None, 0, 255, cv2.NORM_MINMAX)

# Converter para o tipo uint8
log_transformed_img = log_transformed_img.astype(np.uint8)

# Exibir a imagem original e a imagem transformada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(log_transformed_img, cmap='gray')
plt.title('Imagem com Transformação Logarítmica')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('im_cells.png')

# Verificar se a imagem foi carregada corretamente
if img is None:
    print('Não foi possível carregar a imagem.')
    exit()

# Inverter as cores para criar a imagem negativa
negative_img = 255 - img

# Constante de ajuste para controlar o aumento do contraste
c = 1.0

# Aplicar a transformação logarítmica à imagem original
log_transformed_img = c * np.log1p(img.astype(np.float32))
log_transformed_img = cv2.normalize(log_transformed_img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

# Aplicar a transformação logarítmica à imagem negativa
log_transformed_negative_img = c * np.log1p(negative_img.astype(np.float32))
log_transformed_negative_img = cv2.normalize(log_transformed_negative_img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

# Exibir as imagens lado a lado
plt.figure(figsize=(12, 6))

# Exibir a imagem original e a imagem transformada
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(log_transformed_img, cv2.COLOR_BGR2RGB))
plt.title('Imagem com Transformação Logarítmica')
plt.axis('off')

# Exibir a imagem negativa e a imagem negativa com transformação logarítmica
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(negative_img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Negativa')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(log_transformed_negative_img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Negativa com Transformação Logarítmica')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# com for

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('pout.tif', cv2.IMREAD_GRAYSCALE)

# Verificar se a imagem foi carregada corretamente
if img is None:
    print('Não foi possível carregar a imagem.')
    exit()

# Obter as dimensões da imagem
rows, cols = img.shape

# Inicializar o vetor de histograma
hist = np.zeros(256, dtype=int)

# Calcular o histograma
for i in range(rows):
    for j in range(cols):
        pixel_value = img[i, j]
        hist[pixel_value] += 1

# Calcular os centros dos bins para o eixo x
bin_centers = np.arange(256)

# Exibir a imagem
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagem')
plt.axis('off')

# Exibir o histograma em um diagrama de barras
plt.subplot(1, 2, 2)
plt.bar(bin_centers, hist, width=1.0, color='black')
plt.title('Histograma')
plt.xlabel('Níveis de Cinza')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('pout.tif', cv2.IMREAD_GRAYSCALE)

# Verificar se a imagem foi carregada corretamente
if img is None:
    print('Não foi possível carregar a imagem.')
    exit()

# Calcular o histograma
hist, bins = np.histogram(img.flatten(), 256, [0, 256])

# Calcular os centros dos bins para o eixo x
bin_centers = (bins[:-1] + bins[1:]) / 2

# Exibir a imagem
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagem')
plt.axis('off')

# Exibir o histograma em um diagrama de barras
plt.subplot(1, 2, 2)
plt.bar(bin_centers, hist, width=1.0, color='black')
plt.title('Histograma')
plt.xlabel('Níveis de Cinza')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
# Equalize a imagem anterior e mostre novamente seu histograma. Mostra a curva de transformação (consulte https://www.mathworks.com/help/images/histogram-equalization.html)

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('pout.tif', cv2.IMREAD_GRAYSCALE)

# Verificar se a imagem foi carregada corretamente
if img is None:
    print('Não foi possível carregar a imagem.')
    exit()

# Equalizar a imagem
equalized_img = cv2.equalizeHist(img)

# Calcular o histograma da imagem equalizada
hist, bin_edges = np.histogram(equalized_img.ravel(), bins=256, range=[0, 256])

# Calcular os centros dos bins para o eixo x
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Calcular a curva de transformação cumulativa
cdf = hist.cumsum() # CDF(k)=^ksum_i=0 P(i)
cdf_normalized = cdf * hist.max() / cdf.max()

# Exibir a imagem original e a imagem equalizada
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Imagem Equalizada')
plt.axis('off')

# Exibir o histograma da imagem equalizada
plt.subplot(2, 2, 3)
plt.bar(bin_centers, hist, width=1.0, color='black')
plt.title('Histograma Equalizado')
plt.xlabel('Níveis de Cinza')
plt.ylabel('Frequência')

# Exibir a curva de transformação
plt.subplot(2, 2, 4)
plt.plot(bin_centers, cdf_normalized, color='black')
plt.title('Curva de Transformação Cumulativa')
plt.xlabel('Níveis de Cinza')
plt.ylabel('CDF Normalizado')

plt.tight_layout()
plt.show()
```
